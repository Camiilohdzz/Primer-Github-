```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def load_img(path):
    img =cv2.imread(path)
    return img[...,::-1]

img =load_img('20221013_093331.jpg')#*0 + 255


print(f'Alto :{img.shape[0]}')
print(f'Ancho :{img.shape[1]}')
print(f'Canales :{img.shape[2]}')

img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

fig, axs = plt.subplots(1,5, figsize=(20,5))

axs[0].imshow(img,vmax=255,vmin=0)
axs[0].set_title('imagen completa')

#fig_img=axs[1].imshow(img[...,0],vmax=255,vmin=0)
umbral_bajoV = (50,100,100)
umbral_altoV = (80,255,255)
maskV = cv2.inRange(img_hsv, umbral_bajoV, umbral_altoV)
fig_img=axs[1].imshow(maskV, cmap='gray', vmax=255,vmin=0)
axs[1].set_title('canal Green')
fig.colorbar(fig_img,ax=axs[1],shrink=0.5)

umbral_bajoA = (100,100,100)
umbral_altoA = (120,255,255)
maskA = cv2.inRange(img_hsv, umbral_bajoA, umbral_altoA)
fig_img=axs[2].imshow(maskA, cmap='gray', vmax=255,vmin=0)
axs[2].set_title('canal Blue')
fig.colorbar(fig_img,ax=axs[2],shrink=0.5)

umbral_bajoR = (1,100,100)
umbral_altoR = (10,255,255)
maskR = cv2.inRange(img_hsv, umbral_bajoR, umbral_altoR)
fig_img=axs[3].imshow(maskR, cmap='gray', vmax=255,vmin=0)
axs[3].set_title('canal Orange')
fig.colorbar(fig_img,ax=axs[3],shrink=0.5);

umbral_bajoAM = (25,100,100)
umbral_altoAM = (35,255,255)
maskAM = cv2.inRange(img_hsv, umbral_bajoAM, umbral_altoAM)
fig_img=axs[4].imshow(maskAM, cmap='gray', vmax=255,vmin=0)
axs[4].set_title('canal Yellow')
fig.colorbar(fig_img,ax=axs[3],shrink=0.5);
```
